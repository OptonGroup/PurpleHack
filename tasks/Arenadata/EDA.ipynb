```python
df_plan = pd.read_json('telecom100k/telecom100k/plan.json')
df_plan
```

```python
import pandas as pd

# 1) Читаем основной файл с данными PSXStats (df_psx)
# 2) Читаем данные Subscriber (df_subscriber)
# 3) Читаем данные Client (df_client) и оставляем нужные столбцы
# 4) Читаем данные PSXAttrs (df_psxattrs)
# 5) Читаем данные Plan (файл plan.json - df_plan) и делаем его DataFrame


# 1. Объединяем PSXStats с Subscriber:
# Связываем psx.csv (IdSubscriber) и Subscriber (IdOnPSX)
df = pd.merge(df_psx, df_subscriber, left_on='IdSubscriber', right_on='IdOnPSX', how='left')

# 2. Присоединяем данные Client по полю IdClient, которое появилось из Subscriber, и Client.Id:
df = pd.merge(df, df_client, left_on='IdClient', right_on='Id', how='left', suffixes=('', '_client'))

# 3. Присоединяем PSXAttrs по полю IdPSX (psx.csv) и PSXAttrs.Id:
df = pd.merge(df, df_psxattrs, left_on='IdPSX', right_on='Id', how='left', suffixes=('', '_psxattrs'))

# 4. Присоединяем данные Plan, используя IdPlan из Client (уже в объединённом df) и Plan.Id:
df = pd.merge(df, df_plan, left_on='IdPlan', right_on='Id', how='left', suffixes=('', '_plan'))

# Результирующий DataFrame содержит:
# - Данные сессий из psx.csv
# - Все столбцы из Subscriber
# - Два столбца из Client (Id и IdPlan)
# - Все столбцы из PSXAttrs
# - Все столбцы из Plan

df.head(2)
```
